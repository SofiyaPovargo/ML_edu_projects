# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

Для one-hot-encoding'a можно испльзовать комбинацию LabelEncoder + OneHotEncoder (сначала превращаем строчки в числа, а потом числа првращаем в one-hot вектора.)

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline, Pipeline
```

```python
data
```

```python
from sklearn.compose import ColumnTransformer
X = data.drop('Churn', axis=1)
y = data['Churn']
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3, random_state=42)
numeric_cols = X.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = X.select_dtypes(include='object').columns
preprocessor = ColumnTransformer(transformers=[('num', StandardScaler(), numeric_cols), ('cat', OneHotEncoder(), categorical_cols)])
pipeline = Pipeline(steps=[('preprocessor', preprocessor), ('model', LogisticRegression(max_iter=1000))])
```

Выпишите какое лучшее качество и с какими параметрами вам удалось получить

```python
param_grid = {'model__C': [100, 10, 1, 0.1, 0.01, 0.001]}
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='roc_auc', refit=True)
grid_search.fit(X_train, y_train)
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_valid)
roc_auc = roc_auc_score(y_valid, y_pred)

print(f'Лучший параметр C: {grid_search.best_params_}')
print(f'ROC-AUC на тестовой выборке: {roc_auc}')

pred_best_model = best_model
```

# Предсказания

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1]
}

best_score = 0
best_params = {}
for n_est in param_grid['n_estimators']:
    for lr in param_grid['learning_rate']:
        print(f"Текущие параметры: n_estimators={n_est}, learning_rate={lr}")
        
        tuned_model = CatBoostClassifier(
            cat_features=cat_cols,
            n_estimators=n_est,
            learning_rate=lr,
            random_state=42,
            verbose=0
        )
        tuned_model.fit(X_train, y_train)
        
        y_tuned_pred = tuned_model.predict_proba(X_valid)[:, 1]
        roc_auc_tuned = roc_auc_score(y_valid, y_tuned_pred)
        
        if roc_auc_tuned > best_score:
            best_score = roc_auc_tuned
            best_params = {'n_estimators': n_est, 'learning_rate': lr}
            
        print(f"ROC AUC: {roc_auc_tuned}\n")

print(f"Лучшая комбинация параметров: {best_params}")
print(f"Лучшее значение ROC AUC: {best_score}")

X_test = pd.read_csv("C:/Users/Asus/Downloads/test.csv")
submission = pd.read_csv('C:/Users/Asus/Downloads/submission (1).csv')

# Обучение лучшей модели на всём train set
best_model = CatBoostClassifier(**best_params, cat_features=cat_cols, random_state=42, verbose=0)
best_model.fit(X, y)

submission['Churn'] =  best_model.predict_proba(X_test)[:, 1] # используйте best_model.predict_proba(X_test), не забудьте выделить вероятность класса 1.
submission.to_csv('C:/Users/Asus/Downloads/my_submission3.csv', index=False)
```
